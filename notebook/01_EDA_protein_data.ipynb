```python
# --- DATA CLEANING BLOCK ---

# 1. Fix Units: Keep only 'nM' (Nanomolar) to ensure consistency
# We drop rows where units are uM, mM, etc. to avoid conversion errors for now.
df_clean = df[df['standard_units'] == 'nM'].copy()
print(f"Rows after filtering units: {df_clean.shape[0]}")

# 2. Fix Outliers: Remove biologically impossible values
# A reasonable IC50 range is 0.1 nM to 100,000 nM (100 uM).
# Anything above 100,000 nM is basically "inactive".
df_clean = df_clean[df_clean['standard_value'] <= 100000]
df_clean = df_clean[df_clean['standard_value'] > 0] # Remove negatives/zeros
print(f"Rows after removing outliers: {df_clean.shape[0]}")

# 3. Create Target Variable (pIC50)
# pIC50 = 9 - log10(IC50_nM)
import numpy as np
df_clean['pIC50'] = 9 - np.log10(df_clean['standard_value'])

# 4. Drop Metadata Columns (Feature Selection)
cols_to_drop = [
    'molecule_chembl_id', 'target_chembl_id', 'target_name', 
    'bioactivity_type', 'standard_units', 
    'units', 'value', 'bao_label', 'standard_type', 'standard_relation', 'document_chembl_id'
]
# Only drop columns that actually exist in your dataframe
existing_cols_to_drop = [c for c in cols_to_drop if c in df_clean.columns]
df_final = df_clean.drop(columns=existing_cols_to_drop)

print("✅ Data Cleaned! Ready for Modeling.")
df_final.head()
```

Filter Units -> Remove Outliers -> Clean SMILES -> Deduplicate -> Save


```python

import pandas as pd
import numpy as np
from rdkit import Chem
from rdkit.Chem.SaltRemover import SaltRemover

# 1. Reload Data (To recover 'target_name' if it was dropped)
# Make sure path is correct relative to your notebook
df = pd.read_csv("../artifacts/data/master_dataset_with_protein.csv")

# 2. Filter Units & Outliers
# Keep only nM units and remove impossible values
df_clean = df[df['standard_units'] == 'nM'].copy()
df_clean = df_clean[df_clean['standard_value'] <= 100000] # Remove inactive > 100uM
df_clean = df_clean[df_clean['standard_value'] > 0]       # Remove negatives

# 3. Calculate pIC50
df_clean['pIC50'] = 9 - np.log10(df_clean['standard_value'])

# 4. Standardize SMILES (Chemistry Cleaning)
remover = SaltRemover()

def clean_smiles_safe(smiles):
    try:
        mol = Chem.MolFromSmiles(smiles)
        if mol is None: return None
        mol = remover.StripMol(mol) # Remove salts
        return Chem.MolToSmiles(mol, isomericSmiles=False) # Canonicalize
    except:
        return None

print("🧪 Standardizing SMILES (this may take a moment)...")
df_clean['clean_smiles'] = df_clean['smiles'].apply(clean_smiles_safe)
df_clean = df_clean.dropna(subset=['clean_smiles'])

# 5. Handle Duplicates (Now 'target_name' exists!)
# We group by Drug + Protein and take the median potency
print("Dataset size before deduplication:", df_clean.shape[0])

df_final = df_clean.groupby(['clean_smiles', 'target_name'])['pIC50'].median().reset_index()

# 6. Re-attach Protein Features
# We lost them during groupby, so we map them back
protein_features = df_clean[['target_name', 'Protein_Hydrophobicity', 'Protein_Weight', 
                             'Protein_Isoelectric', 'Protein_Aromaticity']].drop_duplicates()

df_ready = pd.merge(df_final, protein_features, on='target_name', how='left')

print(f"✅ SUCCESS: Final Clean Dataset Size: {df_ready.shape}")
print("Columns:", df_ready.columns.tolist())

# 7. Save for the Next Step
df_ready.to_csv('../artifacts/data/clean_training_data.csv', index=False)
print("💾 Saved to artifacts/data/clean_training_data.csv")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# 1. Select only numeric columns for plotting
# We use df_ready (the clean data we just made)
numeric_df = df_ready.select_dtypes(include=[np.number])

# Remove infinite values just in case
numeric_df = numeric_df.replace([np.inf, -np.inf], np.nan).dropna()

print(f"Plotting distributions for {len(numeric_df.columns)} columns...")

# 2. Setup a Grid Layout (3 columns wide)
n_cols = 3
n_rows = (len(numeric_df.columns) + n_cols - 1) // n_cols

plt.figure(figsize=(15, 5 * n_rows))
plt.subplots_adjust(hspace=0.5)

# 3. Loop and Plot
for i, col in enumerate(numeric_df.columns):
    ax = plt.subplot(n_rows, n_cols, i + 1)
    
    # Plot Histogram with KDE
    sns.histplot(data=numeric_df, x=col, kde=True, color='skyblue', edgecolor='black')
    
    plt.title(f"Distribution of {col}")
    plt.xlabel("") # Hide x-label to reduce clutter

plt.show()
```
